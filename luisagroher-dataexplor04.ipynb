```python
xval_errors = []
ndegrees = 10
for ndegree in range(1, ndegrees):
    # each ncol corresponds to a polynomial degree
    X = np.power(X, ndegree)
    y = df["y"].values
    kfold = KFold(len(df), n_folds=10)
    mses = []
    for train, test in kfold:
        # each model is cross validated 10 times
        Xtrain, ytrain, Xtest, ytest = X[train], y[train], X[test], y[test]   
        model = LinearRegression(normalize=True, fit_intercept=False)
        model.fit(X, y)
        ypred = model.predict(Xtest)
        mses.append(mean_squared_error(ypred, ytest))
    xval_errors.append(np.mean(mses))
plt.plot(range(1,ndegrees), xval_errors)
plt.xlabel("degree")
plt.ylabel("xval.error")
```

```python
cv_errors = {}
for r in range(1, len(predictors)):
    for predictor_combos in combinations(predictors, r=r):        
        X = df[list(predictor_combos)].values
        y = df["y"].values
        kfold = KFold(len(df), n_folds=10)
        mses = []
        for train, test in kfold:
            # each model is cross validated 10 times
            Xtrain, ytrain, Xtest, ytest = X[train], y[train], X[test], y[test]   
            model = LinearRegression()
            model.fit(X, y)
            ypred = model.predict(Xtest)
            mses.append(mean_squared_error(ypred, ytest))
        cv_errors[predictor_combos] = np.mean(mses)
```
