```python
import pandas as pd
import sklearn as sk
from sklearn.multiclass import OneVsRestClassifier
from sklearn.multiclass import OneVsOneClassifier
from sklearn.svm import LinearSVC
```

```python
df = pd.read_csv('C:\Users\sk19723\Desktop\setlist.csv')
```

```python
df
```

```python
for col in df.columns[:-1]:
   df[col] = [int(val == True) for val in df[col]]
```

```python
df
```

```python
ndf = df[df.year != 1987]
```

```python
ndf
```

```python
ndf['year'] = [int(str(year)[:3]+'0') for year in ndf['year']]
```

```python
ndf
```

```python
X = ndf.ix[:, :-1]
```

```python
y = ndf['year']
```

```python
clf = OneVsRestClassifier(LinearSVC(random_state=0)).fit(X, y)
```
