```python
import pandas as pd
df = pd.DataFrame.from_csv('https://raw.githubusercontent.com/cmrivers/ebola/master/country_timeseries.csv',
                 index_col=0)
```

```python
#reproduce epidemic curves for each county
df = df.fillna(method = 'bfill') # will fill data from first data point
#df = df.interpolate() # will fill data before data points
df.head()
```

```python
cases_column = []
deaths_column = []
```

```python
for column in df.columns:
    head = column.split('_')
    if head[0] == 'Cases':
        cases_column.append(column)
    if head[0] == 'Deaths':
        deaths_column.append(column)
```

```python
cases = pd.DataFrame(df[cases_column])
cases.plot()
```

```python
deaths = pd.DataFrame(df[deaths_column])
deaths.plot()
```

```python
deaths_liberia = pd.DataFrame(deaths['Deaths_Liberia'])
```

```python
def acf(df):
    import seaborn as sns
    lagged = pd.concat([df.shift(i) for i in range(10)], axis=1)
    lagged.columns = ["lag%s"%i for i in range(10)]
    sns.corrplot(lagged)
```

```python
for c in deaths.columns:
    acf(deaths[c])
    show()
```
